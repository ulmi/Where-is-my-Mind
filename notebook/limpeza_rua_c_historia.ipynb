```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
df = pd.read_excel('./ruas.xlsx', usecols="A", engine='openpyxl')
df.dropna(how='all', inplace=True)
with pd.option_context('display.max_colwidth', None):
  display(df.head(15))
```

```python
df[['nome_rua','descricao']] = df["a coisa"].str.split("\n", 1, expand=True)
```

```python
df[['rua','lugar']] = df["nome_rua"].str.split(" – ", 1, expand=True)
```

```python
df[['mdesc','outra_info']] = df["descricao"].str.split("\nInício: ", 1, expand=True)
```

```python
df[['inicio','fim']] = df["outra_info"].str.split("\nFim: ", 1, expand=True)
```

```python
df.drop(['descricao', 'outra_info'], axis=1, inplace=True)
with pd.option_context('display.max_colwidth', None):
  display(df.head(15))
```

```python
df2=df.replace('\n',' ',regex=True)
with pd.option_context('display.max_colwidth', None):
  display(df2.head(15))
```

```python
df3=df2.replace('Memória descritiva e justificativa: ','',regex=True)
with pd.option_context('display.max_colwidth', None):
  display(df3.head(200))
```

```python
df3.dropna(how='all', inplace=True)
df3.to_csv('teste.csv', encoding='utf-8')
```
